```python
import nltk # nltk procesamiento de lenguaje naturaal

texto = 'Hola, ¿Cómo estas?'
# split conviente en lista cada palabla incluido signos
tokens = texto.split()
print(tokens)
```

```python
texto = 'Hola, ¿Cómo estas?'
# lower conviente mayusculas a minusculas
tokens = texto.lower().split()
print(tokens)
```

```python
texto = 'El gato es negro y el perro es blanco'
# las palablas mas importantes son: gato negro, perro blanco.
# el resto son stop words
# usando nltk
texto = texto.lower()
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('popular') 
nltk.download('all')

stop_words = set(stopwords.words('spanish'))
print(stop_words)
```

```python
tokens = word_tokenize(texto)
print(tokens)
```

```python
# filtrar las palabras mas importantes
texto_filtrado = [word for word in tokens if not word in stop_words]
print(texto_filtrado)
```
