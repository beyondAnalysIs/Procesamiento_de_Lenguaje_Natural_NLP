```python
import nltk # nltk procesamiento de lenguaje naturaal

texto = 'Hola, ¿Cómo estas?'
# split conviente en lista cada palabla incluido signos
tokens = texto.split()
print(tokens)
```

```python
texto = 'Hola, ¿Cómo estas?'
# lower conviente mayusculas a minusculas
tokens = texto.lower().split()
print(tokens)
```

```python
texto = 'El gato es negro y el perro es blanco'
# las palablas mas importantes son: gato negro, perro blanco.
# el resto son stop words
# usando nltk
texto = texto.lower()
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('popular') 
nltk.download('all')

stop_words = set(stopwords.words('spanish'))
print(stop_words)
```

```python
tokens = word_tokenize(texto)
print(tokens)
```

```python
# filtrar las palabras mas importantes
texto_filtrado = [word for word in tokens if not word in stop_words]
print(texto_filtrado)
```

### Stemming y Lemmatization
- stremming. es una tecnica mas simple que elimina los sufijos de las palabras
- lemmatization. es una tecnica mas sofisticada que utiliza la reglas del lenguaje para obtener la base o raiz de una palabra

```python
# importando nltk wordnet. sirve para obtener la raiz de una palabra
import nltk 
nltk.download('wordnet')
```

```python
from nltk.stem import SnowballStemmer
# crear el stemmer en español
stemmer = SnowballStemmer('spanish')
# probando con la palabra 'caminar'
print(stemmer.stem('caminar'))
print(stemmer.stem('caminando'))
print(stemmer.stem('caminó'))
 
```

```python
# importando spacy es_core_news_sm para 
import spacy
# cargar el modelo en español
nlp = spacy.load('es_core_news_sm')
# procesar el texto
doc = nlp('caminar caminando caminó caminará')
# imprimir el lema de cada palabra
for token in doc:
    print(token, '->', token.lemma_)
```

### Recomendaciones de uso
- la lemmatization puede ser mas efectiva que el stemming, pero tambien es mas costosa computacionalmente
- el uso de la lemmatization puede requerir el etiquetado previo
